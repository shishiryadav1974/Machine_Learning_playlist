```python
# this code for solving the above error
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder

# Example: your DataFrame (df)
# df = pd.read_csv("customer_dataset.csv")

# Step 1: Split data
x = df.iloc[:, :-1]
y = df.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Step 2: Replace all weird missing values or mixed types
x_train = x_train.replace(['None', 'none', 'NaN', 'nan', 'NULL', 'null'], np.nan)
x_test = x_test.replace(['None', 'none', 'NaN', 'nan', 'NULL', 'null'], np.nan)

# Step 3: Convert every value to string (OrdinalEncoder expects consistent types)
x_train = x_train.astype(str)
x_test = x_test.astype(str)

# Step 4: Encode safely
oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
x_train_encoded = oe.fit_transform(x_train)
x_test_encoded = oe.transform(x_test)

print("✅ Encoding successful!")
print("x_train_encoded shape:", x_train_encoded.shape)
```
