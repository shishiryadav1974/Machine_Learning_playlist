```python
import pandas as pd
import requests
from bs4 import BeautifulSoup # BeautifulSoup is library which can be used for web scrapping
```

```python
headers={'user-Agent':'MOzilla/5.0 (Windows NT 6.3; Win64) AppleWebkit/537.36 (KHTML,like)'}
webpage=requests.get("https://www.ambitionbox.com/",headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```
